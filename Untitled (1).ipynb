```python

# Import necessary libraries
import pandas as pd

# Load the dataset from an online source (no upload needed)
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"

try:
    # Load dataset
    df = pd.read_csv(url)

    # Display the first few rows
    print("üìå First 5 rows of the dataset:")
    print(df.head())

    # Explore the structure of the dataset
    print("\nüìå Dataset Information:")
    print(df.info())

    # Check for missing values
    print("\nüìå Missing values per column:")
    print(df.isnull().sum())

    # Clean the dataset by dropping missing values (if any)
    df_cleaned = df.dropna()

    print("\n‚úÖ Cleaned dataset (missing values removed if present):")
    print(df_cleaned.head())

except Exception as e:
    print(f"‚ùå An error occurred: {e}")
```

```python
# Compute basic statistics for numerical columns
print("üìä Descriptive Statistics (mean, std, min, max, etc.):")
print(df_cleaned.describe())

# Perform grouping by species and compute the mean of each numerical column
grouped_means = df_cleaned.groupby("species").mean(numeric_only=True)

print("\nüìä Mean values per species:")
print(grouped_means)

# Observations and patterns
print("\nüîç Observations:")
print("- Setosa has the smallest petal and sepal measurements on average.")
print("- Virginica has the largest petal length and width.")
print("- Versicolor is intermediate in most measurements.")
```
